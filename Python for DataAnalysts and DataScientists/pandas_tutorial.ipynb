## Splitting Columns

```python
# Create 2 new columns and then split a column based on the ", "
netflix[["Date Part 1", "Date Part 2"]] = netflix["date_added"].str.split(", ", expand=True)
netflix
```

```python
netflix.rename(columns={"Date Part 2": "Year"}, inplace=True)
netflix
```

```python
# Fill NaN values in the cast column with no_cast
netflix["cast"].fillna(value="no_cast", inplace=True)
netflix
```

## Lambda function

```python
# lambda allows you to write a function in one line.  Below the function the_adder will add x to the number
the_adder = lambda x: x + 210101
the_adder(3535)
```

```python
# The map method will pass each element of a Series(a series)
# is a column of a DataFrame to the lambda function.
# The apply method will pass an entire row to the lambda
#  function.
# Below we are passing each element of the Series
# netflix["listed_in"] to the lambda function as x.
# We then perform the logic with the code after the colon.
# This creates the genre_count and counts the number of values in
# listed_in column separated by ", " 
netflix["genre_count"] = netflix["listed_in"].map(lambda x: len(x.split(", ")))
netflix
```

```python
# Create nonsensical_column that uses the apply method to the row for cast
#  and listed_in and adds the len value together after the ","
netflix["nonsensical_columns"] = netflix.apply(lambda x: len(x["cast"].split(","))
 + len(x["listed_in"].split(",")), axis=1)
netflix
```

## GroupBy, Aggregation, and Sorting Values

```python
# Count the number of movies per country and sort the list
# in descending order
netflix.groupby("country")["show_id"].count()
```

```python
# The above will output a groupby object.  Not very useful.
# We will need to turn the data back into a DataFrame using
# a method called reset_index().
netflix.groupby("country")["show_id"].count().reset_index()
```

```python
# Sort the values by show_id column by country in descending order
netflix_movies_by_country = netflix.groupby("country")["show_id"].count().reset_index().sort_values(by="show_id", ascending=False)
netflix_movies_by_country
```

## Joins(merge) and Unions(concat)

```python
# Import second dataset
second_dataset = pd.read_csv(pwd + "/netflix_titles_second.csv")
```

```python
# Concat the netflix dataset with the second dataset
new_dataset = pd.concat((netflix, second_dataset))
new_dataset
```

```python
# Left join with the netflix_movies_by_country data
netflix_merged = pd.merge(left = new_dataset, right =
netflix_movies_by_country, how="inner", left_on = ["country"], right_on = ["country"])
netflix_merged
```

## Pivot and Melt

```python
# Pivoting data takes it from the "long" format and then
# converts it to a "wide" format
pivot_table = netflix.pivot_table(index="country", columns="type", values="title", aggfunc='count', fill_value=0).reset_index()
pivot_table
```
